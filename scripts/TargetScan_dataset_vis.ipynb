## Visualize TargetScan Data Distribution of 3 primates

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pos_data = {
    'threshold':  ['Top 5%', 'Top 10%', 'Top 15%', 'Top 20%', 'total'],
    'human':      [57500, 100800, 132700, 155600, 228000],
    'chimpanzee': [47600, 84000, 110900, 130000, 188900],
    'mouse':      [34700, 63800, 86600, 104000, 164500]
}

pos_df = pd.DataFrame(pos_data)

neg_data = {
    "threshold": ["Bottom 20%", "Bottom 30%", "Bottom 40%", "Bottom 50%"],
    "human":     [994,       6000,      11764,      19948],
    "chimpanzee":[786,       4600,      9026,       15370],
    "mouse":     [992,       4668,      8645,       15137]
}

neg_df = pd.DataFrame(neg_data)

def plot_stack_barplot(data, title):
    # 准备绘图数据
    thresholds = data['threshold']
    human_vals = data['human']
    chimp_vals = data['chimpanzee']
    mouse_vals = data['mouse']

    # 绘制堆叠柱状图
    plt.figure(figsize=(8, 6))

    # 第一层（human）
    plt.bar(thresholds, human_vals, label='human')
    # 第二层（chimpanzee），底部叠加 human
    plt.bar(thresholds, chimp_vals, bottom=human_vals, label='chimpanzee')
    # 第三层（mouse），底部叠加 human + chimpanzee
    plt.bar(thresholds, mouse_vals, bottom=human_vals + chimp_vals, label='mouse')

    # 设置坐标轴标签和标题
    plt.xlabel('Threshold')
    plt.ylabel('Count of miRNA–Transcript Pairs')
    plt.title(title)

    # 添加图例
    plt.legend()

    # 调整布局后显示
    plt.tight_layout()
    plt.show()
```

```python
plot_stack_barplot(data=pos_df, title="Stacked Barplot of miRNA–Transcript Pairs by Threshold and Species")
plot_stack_barplot(data=neg_df, title="Stacked Barplot of miRNA–Transcript Pairs by Threshold and Species")
```
